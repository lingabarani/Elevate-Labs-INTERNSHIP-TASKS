```python
# ---------------------------------------------
# Exploratory Data Analysis - Titanic Dataset
# ---------------------------------------------

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")

# ---------------------------------------------
# Step 1: Load Dataset
# ---------------------------------------------
# If using Titanic dataset from seaborn
df = sns.load_dataset('titanic')

# Display first few rows
print("First 5 rows of the dataset:")
display(df.head())

# ---------------------------------------------
# Step 2: Basic Information
# ---------------------------------------------
print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
display(df.describe(include='all'))

# Checking for missing values
print("\nMissing Values:")
display(df.isnull().sum())

# ---------------------------------------------
# Step 3: Value Counts for Categorical Columns
# ---------------------------------------------
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
    print(f"\nValue Counts for '{col}':")
    print(df[col].value_counts())

# ---------------------------------------------
# Step 4: Data Visualization
# ---------------------------------------------

# a. Distribution of Age
plt.figure(figsize=(8,5))
sns.histplot(df['age'].dropna(), kde=True, bins=30, color='skyblue')
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# b. Survival Count
plt.figure(figsize=(6,4))
sns.countplot(x='survived', data=df, palette='viridis')
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()

# c. Gender vs Survival
plt.figure(figsize=(6,4))
sns.countplot(x='sex', hue='survived', data=df, palette='coolwarm')
plt.title("Survival by Gender")
plt.show()

# d. Class vs Survival
plt.figure(figsize=(6,4))
sns.countplot(x='class', hue='survived', data=df, palette='Set2')
plt.title("Survival by Passenger Class")
plt.show()

# e. Pairplot for Numerical Relationships
sns.pairplot(df[['age', 'fare', 'pclass', 'survived']].dropna(), hue='survived', palette='husl')
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()

# f. Heatmap for Correlation
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()

# g. Boxplot - Age vs Class
plt.figure(figsize=(8,5))
sns.boxplot(x='class', y='age', data=df, palette='Pastel1')
plt.title("Age Distribution Across Classes")
plt.show()

# h. Scatterplot - Age vs Fare
plt.figure(figsize=(8,5))
sns.scatterplot(x='age', y='fare', hue='survived', data=df, palette='cool')
plt.title("Fare vs Age by Survival Status")
plt.show()

# ---------------------------------------------
# Step 5: Observations
# ---------------------------------------------
print("""
Observations:
1. Most passengers were in 3rd class and younger adults.
2. Female passengers had a higher survival rate than males.
3. Higher-class passengers had a better survival rate.
4. Fare was positively correlated with survival.
5. Some missing values exist in 'age', 'deck', and 'embark_town' columns.
6. Age distribution shows most passengers were between 2040 years.
""")

# ---------------------------------------------
# Step 6: Summary of Findings
# ---------------------------------------------
print("""
Summary:
The Titanic dataset reveals that socioeconomic status and gender significantly affected survival.
Females and passengers in 1st class had a much higher survival rate.
Younger passengers and those who paid higher fares also had better chances of survival.
EDA highlights the importance of feature relationships for predictive modeling.
""")
```
